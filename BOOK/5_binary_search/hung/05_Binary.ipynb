## 프로그래머스

```python
def solution(n, times):
    left, right = 1, max(times) * n  # 최소 시간과 최대 시간 설정

    while left < right:
        mid = (left + right) // 2  # 중간값으로 예상 시간 설정
        total = sum(mid // time for time in times)  # 중간값을 기준으로 심사 가능한 인원 계산

        if total < n:  # 심사할 수 있는 인원수가 대기 인원보다 적은 경우
            left = mid + 1
        else:  # 심사할 수 있는 인원수가 대기 인원보다 많거나 같은 경우
            right = mid

    return left  # 모든 인원을 심사할 수 있는 최소 시간 반환
```
