## Ch 3. Greedy

### 그리디 - 거스름돈


```python
N = int(input())
cnt = 0

money = [500,100,50,10]

for m in money:    
    cnt += N // m
    N = N % m
print(cnt)
```

### 큰수의 법칙
- 다양한 수의 배열에서 전체 M번 더하여 최대의 수를 만드는데
- 같은 수가 연속해서 K번을 초과하면 X


```python
N,M,K = map(int, input().split())
num = list(map(int, input().split()))
num.sort(reverse=True)

total = 0
n_times = M // (K+1)
total += num[0]*n_times*K
total += num[1]*n_times
total += num[0]*(M%(K+1))
print(total)
```

### 숫자 카드 게임

```python
N,M = map(int,input().split())
result = 0

while N!=0:
    num = list(map(int, input().split()))
    min = min(num)
    result = max(result, min)
    N -= 1
print(result)
```

### 1이 될 때까지

```python
N,K = map(int,input().split())
cnt = 0

while N != 1:
    if N % K == 0:
        N = N // K
        cnt += 1
    else:
        N -= 1
        cnt += 1
print(cnt)
```
