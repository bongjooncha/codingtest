## 숫자 카드

```python
def card(x):
    min_in_rows = [min(row) for row in x]
    return max(min_in_rows)

x = [
    [3, 1, 2],
    [4, 1, 4],
    [2, 2, 2]
]

print(card(x))
```

## 1이 될때까지

```python
def sol(N, K):
    operations = 0

    while N >= K:
        while N % K != 0:
            N -= 1
            operations += 1
        N //= K
        operations += 1
    operations += (N - 1)
    return operations

N = 17
K = 4
print(sol(N, K))
```

# 모험가 길드

```python
def solution(N, m):
    # 공포도를 오름차순으로 정렬
    m.sort()
    # 그룹 수를 세기 위한 카운터
    count = 0
    # 현재 그룹에 포함된 모험가의 수
    count2 = 0
    for i in m:
        # 현재 그룹에 해당 모험가를 포함시킴
        count2 += 1
        if count2 >= i:
            # 그룹 결성 성공시 그룹 증가
            count += 1
            count2 = 0 # 초기화

    return count
N = 5
fear_levels = [2, 3, 1, 2, 2]
print(solution(N, fear_levels))
```

# 곱하기 혹은 더하기

```python
def solution(num):
    # 초기값은 첫 번째 문자에 해당하는 숫자
    result = int(num[0])
    # 두 번째 문자부터 마지막 문자까지 확인
    for i in range(1, len(num)):
        # 현재 문자를 숫자로 변환
        n = int(num[i])
        # 현재 수가 1보다 작으면 더하는게 이득, 결과가 0 이거나 1일때도 더하기 연산 수행
        if n <= 1 or result <= 1:
            result += n
        # 그 외의 경우에는 곱하기 연산 수행
        else:
            result *= n

    return result

a = "02984"
print(solution(a))
```

# 문자열 뒤집기

```python
def flip(string):
    # 연속된 숫자를 세기 위해 첫 번째 문자를 현재 그룹으로 설정
    current_group = string[0]
    # 뒤집기 횟수
    flip_count = 0

    # 문자열을 순회하며 연속되지 않은 경우를 세기
    for i in string:
        # 현재 그룹과 다른 문자가 나오면, 뒤집기 횟수를 증가시키고 그룹을 바꿈
        if i != current_group:
            flip_count += 1
            current_group = i

    # 전체를 한 번에 뒤집는 경우가 가능하므로, 반으로 나누어 반환
    return (flip_count + 1) // 2

# 함수 테스트
string = "0001100"
print(flip(string))
```

# 만들 수 없는 금액

```python
def sol(coins):
    # 동전을 오름차순으로 정렬
    coins.sort()

    # 만들 수 있는 최소 금액을 1로 설정
    target = 1
    for i in coins:
        # 만약 현재 동전으로 target 금액을 만들 수 있다면, target 업데이트
        if i <= target:
            target += i
        # 만약 현재 동전이 target보다 크다면, target x
        else:
            break

    # 만들 수 없는 최소 금액 반환
    return target

# 예시 입력
N1 = 5
coins1 = [3, 2, 1, 1, 9]

N2 = 3
coins2 = [3, 5, 7]

# 최솟값 출력
print(sol(coins1))
print(sol(coins2))
```

# 볼링공

```python
def bowling(N, M, weights):
    # 조합의 수를 셀 변수
    count = 0
    # 1부터 M까지의 각 무게에 대해 무게별 볼링공 수를 셈
    ball_counts = [0] * (M + 1)
    for i in weights:
        ball_counts[i] += 1

    # 가능한 모든 무게 조합에 대해서
    for i in range(1, M + 1):
        N -= ball_counts[i] # i 무게의 볼링공을 제외
        count += ball_counts[i] * N # 남은 볼링공과 조합

    return count

N = 5
M = 3
weights = [1, 3, 2, 3, 2]

# 경우의 수 출력
print(bowling(N, M, weights))
```
